```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('movies.csv')
```

```python
movies.head()
```

### What was the runtime for each movie?

```python
ax = movies.plot(x='title', y='runtime(min)', kind="bar", figsize=(15,10), legend=None)
plt.title('Movie runtimes', fontsize = 12)
plt.ylabel("runtime in minutes")
for i in ax.containers:
    ax.bar_label(i,)
    
plt.savefig('runtime.png')    
```

In above graph the runtimes for each movies are shown.

I used a bar chart to get the results

### What movie had the longest runtime?

12 Years a slave has the longest runtime as shown in the graph above.

### Is there any relationship between the length of the movie (runtime) and the number of awards a movie won?

```python
sns.scatterplot(data=movies, x='award_wins', y='runtime(min)')
plt.title('Relationship between movie lenght and award wins', fontsize = 12)
plt.xlabel("number of awards")
plt.ylabel("runtime (min)")
plt.show()

plt.savefig('correlation.png') 
```

There is no relationship between the movie lenght and the number of awards won

To get the result I used a scatterplot

### How many awards and nominations did each movie receive?

```python
ax = movies.plot(x="title", y=['award_wins', 'award_nominations' ], kind="bar", figsize=(15,10))
for i in ax.containers:
    ax.bar_label(i,)
    
plt.savefig('awards.png')     
```

Above graph shows all award wins and nominations. 'the Shape of Water' had the most nominations. 'Parasite' had the most wins.

To get the results I used a bar chart
